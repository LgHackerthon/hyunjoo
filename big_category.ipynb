```python
# !pip install prophet
from prophet import Prophet
```

```python

m_df_1 = Prophet(seasonality_mode = 'additive',
                changepoint_prior_scale=0.1,
                yearly_seasonality=20,
                weekly_seasonality=20,
                daily_seasonality=20)
```

```python

m_df_2 = Prophet(seasonality_mode = 'additive',
                changepoint_prior_scale=0.1,
                yearly_seasonality=20,
                weekly_seasonality=20,
                daily_seasonality=20)


m_df_3 = Prophet(seasonality_mode = 'additive',
                changepoint_prior_scale=0.1,
                yearly_seasonality=20,
                weekly_seasonality=20,
                daily_seasonality=20)


m_df_4 = Prophet(seasonality_mode = 'additive',
                changepoint_prior_scale=0.1,
                yearly_seasonality=20,
                weekly_seasonality=20,
                daily_seasonality=20)


m_df_5 = Prophet(seasonality_mode = 'additive',
                changepoint_prior_scale=0.1,
                yearly_seasonality=20,
                weekly_seasonality=20,
                daily_seasonality=20)
```

```python
m_df_1.fit(df_1)
m_df_2.fit(df_2)
m_df_3.fit(df_3)
m_df_4.fit(df_4)
m_df_5.fit(df_5)
```

```python
future = m_df_1.make_future_dataframe(periods=60)
forecast0 = m_df_1.predict(future)
future = m_df_2.make_future_dataframe(periods=60)
forecast1 = m_df_2.predict(future)
future = m_df_3.make_future_dataframe(periods=60)
forecast2 = m_df_3.predict(future)
future = m_df_4.make_future_dataframe(periods=60)
forecast3 = m_df_4.predict(future)
```
