```python
def check_missing_col(dataframe):
    missing_col = []
    for col in dataframe.columns:
        missing_values = sum(dataframe[col].isna())
        is_missing = True if missing_values >= 1 else False
        if is_missing:
            print(f'결측치가 있는 컬럼은: {col} 입니다')
            print(f'해당 컬럼에 총 {missing_values} 개의 결측치가 존재합니다.')
            missing_col.append([col, dataframe[col].dtype])
    if missing_col == []:
        print('결측치가 존재하지 않습니다')
    return missing_col

missing_col = check_missing_col(train)
```

```python
df=train.drop(['ID','제품'],axis=1)
df=pd.melt(df,id_vars=['대분류','중분류','소분류','브랜드'],var_name='date_time')
```

```python
categorical = ['대분류','중분류','소분류','브랜드']

quantitative = ['value','year','month','day']

print('카테고리형(정성적) columns : {0} 개'.format(len(categorical)))
print('수치형(정량적) columns : {0} 개'.format(len(quantitative)))
```

```python
fig, axes = plt.subplots(2,2, figsize=(70,70))
fig.suptitle('Distribution of customers per category', fontsize=100)

for ax, feature in zip(axes.flatten(), categorical):
    sns.countplot(data = df_1, x = feature, ax=ax)
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(20,20))
fig.suptitle('Distribution of quantitative features', fontsize=40)
#plt.tight_layout()

for ax,feature in zip(axes.flatten(),quantitative):
    sns.boxenplot(data = df_1, x = feature, ax=ax, color='#f15354')
plt.show()
```

```python
from scipy import stats
def IQR(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 -(1.5 * IQR)
    upper_bound = Q3 +(1.5 * IQR)
    return lower_bound, upper_bound

outliers = df_1[quantitative].apply(lambda column: IQR(column))
print('Range of outliers by method')
outliers
```

```python
def IQRsum(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 -(1.5 * IQR)
    upper_bound = Q3 +(1.5 * IQR)
    return ((column < lower_bound) | (column > upper_bound)).sum()

outliers = df_1[quantitative].drop(['year', 'month', 'day'], axis = 1).apply(lambda column: IQRsum(column))
print('Number of outliers by method')
outliers
```

```python
# 수치형 데이터 상관관계 히트맵 시각화
from sklearn.preprocessing import MinMaxScaler
train_corr = df_1[quantitative]
scaler= MinMaxScaler() 
df_1[train_corr.columns] = scaler.fit_transform(train_corr[train_corr.columns])
corr28 = train_corr.corr(method= 'pearson')

plt.figure(figsize=(12,10))
sns.heatmap(data = corr28, annot=True, fmt = '.2f', linewidths=.5, cmap='Blues')
plt.title('Correlation between features', fontsize=30)
```
