```python
df_big=df_big.T
df_big.head()
```

```python
Q1=df_big[list(df_big.columns)].quantile(q=0.25)
Q3=df_big[list(df_big.columns)].quantile(q=0.75)
print(Q1)
print(Q3)
```

```python
IQR=Q3-Q1
IQR
```

```python
#이상치 제거

for col in df_big.columns:
  lower_bound = Q1[col] - (1.5 * IQR[col])
  upper_bound = Q3[col] + (1.5 * IQR[col])
  df_big[col] = df_big[col].apply(lambda x: lower_bound if x <= lower_bound else x)
  df_big[col] = df_big[col].apply(lambda x: upper_bound if x >= upper_bound else x)
```

```python
sns.boxplot(data=df_big)
```
